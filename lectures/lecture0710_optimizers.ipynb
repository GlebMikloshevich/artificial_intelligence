## Gradient descent algoritm

### linear gradient descent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
a = 0.01
w0 = 10

def f(w):
    return 10*(w-7)**2

def df(w):
    return 20*(w-7)


steps = 0
while abs(w0 - 7) > 0.0001:
    w0 -= df(w0)*a
    steps += 1
    
    
print(f'steps: {steps}')
print(f'w0: {w0}')
print(f'min of f(w0): {f(w0)}')
```

### two-dimensional gradient descent
calculate vector-gradient of f(x)

```python
w0 = 10
b0 = 9
a = 0.01


def f(w, b):
    return 10*(w-7)**2 + 7*(b-5)**2
# min f(w, b) == f(7, 5)

def df_dw(w):
    return 20*(w-7) 

def df_db(b):
    return 14*(b-5)


steps = 0
while abs(w0 - 7) > 0.0001 or abs(b0 - 5) > 0.0001:
    w0 -= df_dw(w0)*a
    b0 -= df_db(b0)*a
    steps += 1
    
    
print(f'steps: {steps}')
print(f'w0: {w0}')
print(f'b0: {b0}')
print(f'min of f(w0, b0): {f(w0, b0)}')
```

## Monte Carlo method

```python
def f(w):
    return 10*(w-7)**2

def df(w):
    return 20*(w-7)

start = 6.5
end = 7.5
min_value = 1e100
val = 1e100
accuracy = 0.001
steps = 0
random.seed(0)

while abs(val-7) > accuracy:
    val = start + random.random()
    if f(val) < min_value:
        min_value = val
    steps += 1
    
print(steps)
```
